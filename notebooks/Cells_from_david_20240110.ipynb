```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import healpy as hp
import os
import sys
```

```python
with np.load('/scratch/aew492/quasars/quaia_PS/cl_gg_bf_G20.0_1bin.npz') as res:
    ell = res['ell']
    cl_gg = res['cl_gg']
```

```python
ell.shape, cl_gg.shape
```

```python
# plot
fig, ax = plt.subplots(figsize=(8,4))
ax.plot(ell, cl_gg, 'b-', alpha=0.8)
ax.loglog()
ax.set_xlabel(r'$\ell$')
ax.set_ylabel(r'$C_\ell^{gg}$')
ax.grid(lw=0.5, alpha=0.5)
ax.set_title('Quaia G=20.0 best-fit $C(\ell)$')
```

```python
# match axis limits and compare to Fig 5 in Quaia x CMB paper
fig, ax = plt.subplots(figsize=(8,4))
ax.plot(ell, cl_gg, 'b-', alpha=0.8)
ax.axvline(45, c='k', alpha=0.5)
ax.loglog()
ax.set_xlim(1e1,1e3)
ax.set_ylim(3e-9,4e-5)
ax.set_xlabel(r'$\ell$')
ax.set_ylabel(r'$C_\ell^{gg}$')
ax.grid(lw=0.5, alpha=0.5)
ax.set_title('Quaia G=20.0 best-fit $C(\ell)$')
```
